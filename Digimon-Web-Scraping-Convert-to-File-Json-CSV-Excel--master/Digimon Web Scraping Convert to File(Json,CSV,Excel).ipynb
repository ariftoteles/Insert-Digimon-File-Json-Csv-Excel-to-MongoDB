```python
from bs4 import BeautifulSoup
import requests

url = 'http://digidb.io/digimon-list/'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')

tbody = soup.find('tbody') #find body
head = ['#','Image','Digimon','Stage','Type','Attribute','Memory','Equip Slots','HP','SP','Atk','Def','Int','Spd']

image = []
img = tbody.find_all('img', style="vertical-align:middle;")
for k in img:
    image.append(k['src'])

body2 = []  
for i in range (len(tbody.find_all('tr'))): 
    tr = tbody.find_all('tr')[i]  
    body = [] 
    for j in range(len(tr.find_all('td'))): 
        body.append((tr.find_all('td')[j].text).replace('\xa0','').replace(' ','')) 
    body2.append(body)
data = []
data.append(head)

for i in range(len(body2)): 
    x = image[i] 
    y = body2[i] 
    y.insert(1,x) 
    data.append(y)
data
```

```python
# Write to excel

import xlsxwriter

book = xlsxwriter.Workbook('Digimon.xlsx')
sheet = book.add_worksheet()

row = 0 
for i in data:
    for j in  range(len(i)):
        sheet.write(row,j,i[j])
    row += 1
book.close()
```

```python
## Write to csv
## ubah list of list ke list of dictionary
import csv

outDict = []
for i in data:
    outDict.append(dict(zip(data[0], i)))##ZIP(pertama,dua) pertama tambah dua
outDict

with open('DigimonCsv.csv','a',newline='') as mycsv:
    writer = csv.DictWriter(mycsv, delimiter = ',',fieldnames=['#','Image',
 'Digimon',
 'Stage',
 'Type',
 'Attribute',
 'Memory',
 'Equip Slots',
 'HP',
 'SP',
 'Atk',
 'Def',
 'Int',
 'Spd'])
    writer.writerows(outDict)
```

```python
## Write Json
import json
outDict = outDict[1:]
outDict
with open('DigimonJson.json','w') as myJson:
    json.dump(outDict, myJson)
```
